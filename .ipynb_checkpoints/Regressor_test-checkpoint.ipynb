```python
from flaregress.io import DatabaseHandler
from flaregress.io import GOESListHandler, GOESXrayRetriever,LASPf107Retriever, DatabaseHandler
from flaregress.regress import LinearRegressor
import numpy as np
import matplotlib.pyplot as plt
```

Loads data from the sample databases and runs linear regression on that.

```python
train_x, train_y = np.load('./data/sample_databases/train_x.npy'), np.load('./data/sample_databases/train_y.npy')
test_x, test_y = np.load('./data/sample_databases/test_x.npy'), np.load('./data/sample_databases/test_y.npy')
lr = LinearRegressor()
lr.fit(train_x, train_y)
predictions = lr.predict(test_x)
print('RMSE:', lr.rmse(test_y, predictions))
```

```python
plt.plot(predictions)
plt.plot(test_y)
plt.show()
```
